```python
import glob
import os.path

import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
plt.rcParams['figure.figsize'] = (18, 10)

field = img.imread('images/field.png')
```

```python
chances_files = glob.glob(os.path.join('data', 'StrataData_Chances_*.csv'))
chances = pd.concat([pd.read_csv(f, index_col=0) for f in chances_files])

# excluding penalty kicks
chances = chances[(chances.location_x != '-') & (chances.location_y != '-')]
chances.location_x = pd.to_numeric(chances.location_x)
chances.location_y = pd.to_numeric(chances.location_y)

plt.imshow(field, extent=[0, 420, -136, 136])
plt.plot(chances.location_y, chances.location_x, ls='', marker='o')
```
